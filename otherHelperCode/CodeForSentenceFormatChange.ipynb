# to flatten the actors

```python
for record in sen.find():
    listnew=list(set([item for sublist in record['actor'] for item in sublist]))
    print(listnew)
    sen.update_one(
    {
         "_id": record['_id']   #haha, great now it is working, need to use _id
        }, 
      {
                "$set": {
                    "actor": listnew
                }
            })
sen.commit()
```
